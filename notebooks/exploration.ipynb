# üîç Explorace featur a indik√°tor≈Ø

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.rcParams['figure.figsize'] = (14, 6)
sns.set(style="whitegrid")
```

```python
# Naƒçten√≠ featur z processed slo≈æky
df = pd.read_parquet("../data/processed/featurized_data.parquet")
df.head()
```

```python
# Korelace mezi featurami
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, cmap="coolwarm", center=0)
plt.title("üìà Korelace featur")
plt.show()
```

```python
# Uk√°zka vizualizace MACD
df[['close', 'macd', 'macd_signal']].plot(title="MACD a zav√≠rac√≠ cena")
plt.show()
```

```python
# RSI vizualizace
df['rsi_14'].plot(title="RSI (14)")
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.show()
```

```python
# Breakouty
df['close'].plot(label='close', alpha=0.6)
df[df['breakout_high']].close.plot(style='^', color='green', label='breakout high')
df[df['breakout_low']].close.plot(style='v', color='red', label='breakout low')
plt.title("Detekovan√© breakouty")
plt.legend()
plt.show()
```
